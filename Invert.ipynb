```python
import re
import os
```

```python
# Node and Linked List Class Implementation

class node:
    
    def __init__(self,term):
        self.term = term
        self.next = None
        self.freq = 0
        self.postings = {}
        
    def update_node(self,doc,position):
        self.postings.setdefault(doc, [])
        self.postings[doc].append(position)
        self.freq = len(self.postings)

class linkList:
    
    def __init__(self):
        self.head = None
    
    def insert_node(self,term):
        new_node = node(term)
        
        if self.head is None:
            self.head = new_node
        
        else:
            last_node = self.head
            
            while last_node.next:
                  last_node = last_node.next
                    
            last_node.next = new_node
            
    def print_list(self):
        curr_node = self.head
        
        while curr_node:
            print("Term: " , curr_node.term , " " , "Frequency: " , curr_node.freq , " " , "Postings List: " , curr_node.postings)
            curr_node = curr_node.next
            
    def check_len(self):
        curr_node = self.head
        count = 0
        
        while curr_node:
            curr_node = curr_node.next
            count = count + 1
            
        return count
    
    def check_existence(self,term):
        curr_node = self.head
        check = False
        
        while curr_node:
            
            if curr_node.term == term:
                check = True
                break
                
            else:
                curr_node = curr_node.next
                
        return check
    
    def update_list(self,term,doc,position):
        
        curr_node = self.head
       
        
        while curr_node:
            
            if curr_node.term == term:
                curr_node.update_node(doc,position)
                break
            
            else:
                curr_node = curr_node.next
                
    def write_terms_to_file(self):
        
        curr_node = self.head
        file = open('Terms.txt', mode='w', encoding = 'utf-8')
       
        while curr_node:
            file.write(curr_node.term)
            file.write(":")
            file.write(str(curr_node.freq))
            file.write("\n")
            curr_node = curr_node.next
        
        file.close()
        
    def write_postings_to_file(self):
        
        curr_node = self.head
        file = open('Postings.txt', mode='w', encoding = 'utf-8')
       
        while curr_node:
            
            for doc,positions in curr_node.postings.items():
                
                file.write(doc)
                file.write(":")
                
                for index,position in enumerate(positions):
                    
                    if index == len(positions) - 1:
                        file.write(str(position))
                        file.write(";")
                        
                    else:
                        file.write(str(position))
                        file.write(",")
                        
            file.write("\n")
            curr_node = curr_node.next
        
        file.close()
        
        
        
```

```python
# Pre-processing Stop Words

file = open('Closed Class Word List.txt', encoding = 'utf-8')
stopwords = []

for stopword in file.read().split():
    stripped_stopword = stopword.strip()
    clean_stopword = re.sub(r'[\d+\ufeff\u0621\u200ca-zA-Z\u060C\u066B\u066C\u06D4\(\)\"\u2018\u2019\u202a\u202b\u202c\u06EC\',./]','', stripped_stopword)
    if clean_stopword != '':
        stopwords.append(clean_stopword)
                    
                    
stopwords_list = list(set(stopwords))
print(len(stopwords_list))

print(stopwords_list)
```

```python
# Pre-processing the Urdu Corpus 

directory = 'Urdu Corpus/'
files = os.listdir(directory)
words = []

for file_name in files:
   
    file_path = os.path.join(directory, file_name)
    
    if os.path.isfile(file_path):
        with open(file_path,encoding='utf-8') as file:
            content = file.read()
            
            for word in content.split():
                stripped_word = word.strip()
                clean_word = re.sub(r'[\d+\ufeff\u200c\u0621a-zA-Z\u060C\u066B\u066C\u06D4\(\)\"\u2018\u2019\u202a\u202b\u202c\u06EC\',./]','', stripped_word)
                if clean_word != '':
                    words.append(clean_word)
            

                                 
words_list = list(set(words))
print(len(words_list))

print(words_list)
    
```

```python
# Removing Stop Words from the dictionary and sorting the dictionary

words_list = [word for word in words_list if word not in stopwords_list]
sorted_words_list = sorted(words_list)
print(sorted_words_list)

       
```

```python
# Constructing the Inverted Index

inverted_index = linkList()

for word in sorted_words_list:
    inverted_index.insert_node(word)
    
print(inverted_index.check_len())
```

```python
# Adding document ids and positions to the postings lists

directory = 'Urdu Corpus/'
files = os.listdir(directory)


for file_name in files:

    file_path = os.path.join(directory, file_name)

    if os.path.isfile(file_path):
        with open(file_path,encoding='utf-8') as file:
            content = file.read()
            filtered_words = [word for word in content.split() if word != '' and word != ' ']
            
            for position,word in enumerate(filtered_words):
                stripped_word = word.strip()
                clean_word = re.sub(r'[\d+\ufeff\u200c\u0621a-zA-Z\u060C\u066B\u066C\u06D4\(\)\"\u2018\u2019\u202a\u202b\u202c\u06EC\',./]','', stripped_word)
                if inverted_index.check_existence(clean_word):
                    inverted_index.update_list(clean_word,file_name,position)


inverted_index.print_list()
```

```python
# Writing the Index and Postings files to disk

inverted_index.write_terms_to_file()
inverted_index.write_postings_to_file()
```
